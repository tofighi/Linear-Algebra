# Calculating the cross product of vectors

```python
import numpy as np

def unit_vector(vector):
    """ Returns the unit vector of the vector.  """
    return vector / np.linalg.norm(vector)

def radian_angle_between(v1, v2):
    """ Returns the angle in radians between vectors 'v1' and 'v2'::
    """
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))

#initialize arrays
A = np.array([4, -2, 3])
B = np.array([0, 3, 1])



#compute cross product A,B
C = np.cross(A, B)

print(C)
print(unit_vector(C))
print(radian_angle_between(A,B))
```

```python
magnitude = np.linalg.norm(C)
print(magnitude)
```
